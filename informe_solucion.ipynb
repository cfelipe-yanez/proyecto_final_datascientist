# Informe final de la solución

### Pasos del plan realizados y pasos omitidos

Se realizaron todos los pasos principales definidos en el plan de trabajo.
En primer lugar, se cargaron y unificaron los datos provenientes de distintas fuentes, asegurando la consistencia mediante el uso del identificador customerID. Posteriormente, se llevó a cabo un análisis exploratorio de datos (EDA) para comprender la distribución de la variable objetivo, analizar variables numéricas y categóricas, y detectar posibles problemas de calidad de datos.

Luego, se realizó el preprocesamiento de los datos, que incluyó la creación de la variable objetivo (churn), la conversión de variables numéricas, el manejo de valores faltantes y la codificación de variables categóricas. Finalmente, se entrenaron y evaluaron dos modelos de clasificación: Regresión Logística y Random Forest.

No se realizaron ajustes avanzados de hiperparámetros ni validación temporal. Estos pasos fueron omitidos para mantener el enfoque en una solución clara, reproducible y alineada con el alcance del proyecto, ya que los modelos base alcanzaron un nivel de calidad satisfactorio según los criterios de evaluación.

### Dificultades encontradas y cómo se resolvieron

Una de las principales dificultades fue el manejo de variables con tipos de datos incorrectos, en particular la columna TotalCharges, que inicialmente se encontraba como texto. Este problema se resolvió convirtiendo la columna a formato numérico y eliminando las filas con valores faltantes resultantes.

Otra dificultad importante fue la obtención de métricas irreales (AUC-ROC y accuracy iguales a 1.0), lo cual indicaba un problema de data leakage. Este error se corrigió al revisar cuidadosamente el flujo de preprocesamiento, asegurando que la variable objetivo y columnas derivadas no fueran incluidas dentro de las variables predictoras, y aplicando correctamente la codificación de variables antes de la división en conjuntos de entrenamiento y prueba.

### Pasos clave para resolver la tarea

Los pasos más importantes para lograr una solución correcta fueron:

- La correcta definición de la variable objetivo a partir de la columna EndDate.

- La limpieza y preparación de los datos, especialmente la conversión de variables numéricas y el tratamiento de valores faltantes.

- La codificación adecuada de variables categóricas mediante one-hot encoding.

- La evaluación de los modelos utilizando la métrica AUC-ROC, que permite medir la capacidad de discriminación entre clientes que cancelan y los que permanecen.


### Modelo final y nivel de calidad

El modelo final seleccionado fue Random Forest, ya que obtuvo el mejor desempeño entre los modelos evaluados, con un AUC-ROC de 0.876 y una exactitud de 0.847.
De acuerdo con los criterios de evaluación del proyecto, este resultado corresponde a un nivel de calidad alto, superando el umbral requerido para una solución efectiva y confiable.

Esto facilita la implementación de estrategias de retención dirigidas, como la oferta de promociones, descuentos o planes personalizados, reduciendo pérdidas de clientes y optimizando los recursos del equipo de marketing.
